```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Datos
X = np.array([35, 50, 70, 85, 90]).reshape(-1, 1)
y = np.array([0, 0, 1, 1, 1])

# Crear el modelo de regresión logística
model = LogisticRegression()

# Ajustar el modelo a los datos
model.fit(X, y)

# Hacer predicciones
puntuaciones = np.linspace(20, 100, 100).reshape(-1, 1)
probabilidades = model.predict_proba(puntuaciones)[:, 1]

# Graficar los datos y la función sigmoide ajustada
plt.scatter(X, y, color='red', label='Datos')
plt.plot(puntuaciones, probabilidades, label='Modelo de Regresión Logística')
plt.xlabel('Puntuación en el Examen')
plt.ylabel('Probabilidad de Admisión')
plt.title('Probabilidad de Admisión según la Puntuación')
plt.legend()
plt.grid(True)
plt.show()
```
